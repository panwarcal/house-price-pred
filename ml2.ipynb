```python
import pandas as pd
import numpy as np
```

```python
amz_rw = pd.read_csv('DS - Assignment Part 2 data set/amz_com-ecommerce_sample.csv', encoding='unicode_escape')
print(amz_rw.shape)
amz_rw.head()
```

```python
flp_rw = pd.read_csv('DS - Assignment Part 2 data set/flipkart_com-ecommerce_sample.csv')
print(flp_rw.shape)
flp_rw.head()
```

```python
rw = pd.merge(amz_rw, flp_rw, on=('uniq_id'), \
    suffixes=('_amz', '_flp'))
```

```python
print(rw.shape)
rw.head()
```
